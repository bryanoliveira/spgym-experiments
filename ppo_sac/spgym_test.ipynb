```python
from PIL import Image
import sliding_puzzles
```

```python
env = sliding_puzzles.make(
    render_mode="human",
    # w=2,
    # h=2,
    # sparse_rewards=False,
    # # move_reward=-1,
    # shuffle_steps=8,
    # # shuffle_target_reward=-0.5,
    # render_shuffling=True,
    # variation=["raw", "onehot", "image"][0],
    # image_folder=[
    #     "/mnt/data/Documents/sliding-puzzle/imgs/single",
    #     "/mnt/data/Documents/sliding-puzzle/imgs/mnist",
    #     "/mnt/data/Documents/sliding-puzzle/imgs/imagenet-1k",
    # ][1],
    # # background_color_rgb=(255, 0, 0)
    # # circular_actions=True,
    h=2,
    image_background=[0, 0, 0],
    image_folder="/mnt/data/Documents/sliding-puzzle/imgs/single",
    image_size=(84, 84),
    invalid_move_reward=-1,
    shuffle_steps=8,
    sparse_mode="win",
    sparse_rewards=False,
    variation="image",
    w=2,
    win_reward=10,
)


def act(action):
    obs, reward, done, trunc, info = env.step(
        env.action_meanings.index(action)
    )
    print(reward, done, trunc, info)
    return Image.fromarray(obs)
```

```python
obs, _ = env.reset()
print(env.state)
Image.fromarray(env.render(mode="rgb_array"))
```
